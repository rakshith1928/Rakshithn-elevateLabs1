```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ['Age', 'Fare']

plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age')
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x=df['Fare'])
plt.title('Boxplot of Fare')
plt.show()
```

```python
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] >= lower) & (df[col] <= upper)]

df_no_outliers = remove_outliers(df, 'Age')
df_no_outliers = remove_outliers(df, 'Fare')
```

```python
df_no_outliers
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df_no_outliers['Age'])
plt.title('Age after removing outliers')
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x=df_no_outliers['Fare'])
plt.title('Fare after removing outliers')
plt.show()

```
